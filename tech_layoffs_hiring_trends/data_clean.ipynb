# Tech Workforce Intelligence Analytics: Hiring, Layoffs, AI Adoption & Employee Sentiment Analysis

* This project focuses on analyzing workforce trends in the global technology industry using data related to hiring patterns, layoffs, AI adoption, employee sentiment, and market conditions.

* The objective is to identify workforce trends, understand factors influencing organizational decisions, and generate actionable business insights. Using SQL for business queries, Python for exploratory and advanced analysis, and Power BI for interactive dashboards, the project aims to uncover meaningful patterns and support data-driven decision-making through workforce intelligence analysis.

```python
from sqlalchemy import create_engine, text
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
engine = create_engine(
    "mysql+mysqlconnector://root:123456@localhost:3306/tech_layoffs_hiring_trends"
)

with engine.connect() as conn:
    result = conn.execute(text("SHOW TABLES"))
    tables = [row[0] for row in result]

print("Total Tables:", len(tables))
print("Table Names:")
for table in tables:
    print("-", table)
```

```python
tables = ["tech_layoffs_hiring_trends"]
for table in tables:
       print(f"\n Table: {table}")
       query = text(f"SELECT COUNT(*) FROM {table}")
       df = pd.read_sql_query(query, engine)
       print(f"{table}", df.iloc[0,0])
       display(pd.read_sql(f"SELECT * FROM {table} LIMIT 5", engine))
```

```python
df = pd.read_sql_query(text("""select * from tech_layoffs_hiring_trends"""),engine)
df
```

```python
print(f"data info : {df.info()}")
print("-"* 40)
print(f" null check : \n{df.isnull().sum()}")
print("-"* 40)
print(f"duplicate data : {df.duplicated().sum()}")
```

```python
# remove whitespace from object columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()
```

```python
df = df.drop(columns=['record_id'])
```

```python
df
```

```python
df.to_sql(name="clean_dataset",
          con=engine , 
          if_exists="replace" , 
          index=False,
          chunksize=500)
print("clean dataset save in mysql")
```

```python
df_verify = pd.read_sql("SELECT * FROM clean_dataset LIMIT 5", engine)
display(df_verify)
```
