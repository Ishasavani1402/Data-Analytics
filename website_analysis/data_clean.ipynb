```python
import pandas as pd
df = pd.read_csv("website_analysis.csv" ,skiprows=1)
display(df.head(5))
```

```python
df = df.rename(columns={
    "Session primary channel group (Default channel group)" : "chenal_group",
    "Date + hour (YYYYMMDDHH)" : "date_hour",
    "Users" : "user",
    "Sessions" : "session",
    "Engaged sessions" : "engaged_session",
    "Average engagement time per session" : "avg_eng_time_per_session" ,
    "Engaged sessions per user" : "engaged_session_per_user",
    "Events per session" : "event_per_session",
    "Engagement rate" : "engagement_rate" , 
    "Event count" : "event_count"
})
df
```

```python
print(df.info())
print("-"*20)
print(f"null check : \n{df.isnull().sum()}")
print("-"*20)
print(f"duplicate check : {df.duplicated().sum()}")
print("-"*20)
print(f"data shape :{df.shape}")
```

```python
# Convert column to string first
df['date_hour'] = df['date_hour'].astype(str)

# Extract Date part (first 8 digits)
df['date'] = pd.to_datetime(
    df['date_hour'].str[:8],
    format='%Y%m%d'
)

# Extract Hour part (last 2 digits)
df['hour'] = df['date_hour'].str[8:10].astype(int)
```

```python
df.info()
```

```python
for i in df.select_dtypes(include="object"):
    print(f"{i} : \n{df[i].unique()}")
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.to_csv('clean_data.csv' , index=False)
```
